```python
import pandas as pd

data = {
    "Question": [
        "How much is the admission fee?",
        "How can I apply for a hostel?",
        "When will exams start?",
        "What is the last date for admission?",
        "Where can I get the timetable?",
        "How to contact the admin office?",
        "What are the library timings?",
        "Is there any scholarship available?"
    ],
    "Answer": [
        "The admission fee is â‚¹5000.",
        "You can apply for a hostel by filling the form online at hostel.university.edu.",
        "Exams will begin in December as per the academic calendar.",
        "The last date for admission is 30th August.",
        "You can download the timetable from the university website under the academics section.",
        "You can contact the admin office at admin@university.edu or call +91-9876543210.",
        "The library is open from 9 AM to 7 PM from Monday to Saturday.",
        "Yes, scholarships are available based on merit. Visit scholarship.university.edu."
    ]
}

df = pd.DataFrame(data)
df.to_csv("faq.csv", index=False)
df
```

```python
import nltk
nltk.download('punkt_tab')

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load dataset
df = pd.read_csv("faq.csv")

# Text Preprocessing
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

def preprocess(text):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(w) for w in tokens if w.isalnum() and w not in stop_words]
    return " ".join(tokens)

# Preprocess all questions
df["Processed"] = df["Question"].apply(preprocess)

# TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["Processed"])
```

```python
def get_answer(user_query):
    processed_query = preprocess(user_query)
    query_vec = vectorizer.transform([processed_query])

    similarities = cosine_similarity(query_vec, X).flatten()
    index = similarities.argmax()

    if similarities[index] < 0.2:  # threshold to avoid random answers
        return "Sorry, I could not find an answer for that."

    return df.iloc[index]["Answer"]

# Test
print(get_answer("When are exams?"))
print(get_answer("How to join hostel?"))
```

```python
print("ðŸŽ“ University FAQ Chatbot Ready! Ask your questions. Type 'exit' to stop.\n")

while True:
    query = input("You: ")
    if query.lower() == "exit":
        print("Chatbot: Goodbye ðŸ‘‹")
        break

    answer = get_answer(query)
    print("Chatbot:", answer, "\n")
```
